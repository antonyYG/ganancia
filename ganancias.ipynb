```python
# PRIMERO: Verifica que estás usando la versión correcta
import tensorflow as tf
print(f"TensorFlow: {tf.__version__}")
print(f"Keras: {tf.keras.__version__}")

# Tus datos
import pandas as pd
import numpy as np

a = int(input("¿Cuántas fechas va a ingresar? "))
registros = []
for i in range(1, a + 1):
    fecha = input(f"Ingrese el año {i}: ")
    ganancia = float(input(f"Ingrese la ganancia del año {fecha}: "))
    registros.append((fecha, ganancia))

df = pd.DataFrame(registros, columns=["Fecha", "Ganancia"])
df['estado'] = np.where(df['Ganancia'] >= 1000, 1, 0)

X = df['Ganancia'].values.reshape(-1, 1)
y = df['estado'].values

# Crear modelo
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=1, validation_split=0.2, batch_size=8)
```

```python
pesos = []
for layer in model.layers:
    pesos_capa = []
    for peso in layer.get_weights():
        pesos_capa.append(peso.tolist())
    pesos.append(pesos_capa)

# Guardar como JSON
import json
with open('modelo_pesos.json', 'w') as f:
    json.dump(pesos, f)

print("✅ Pesos exportados para JavaScript!")
```
